```python
import pandas as pd

# 載入數據
df1 = pd.read_excel(r'C:\Users\User\Dropbox\Movie\鄭伊廷\台灣國片長片資料.xlsx')
df2 = pd.read_excel(r'C:\Users\User\Dropbox\Movie\游兆愷\電影資料_基本變數.xlsx')
```

```python
# df1 和 df2 是已經載入的 DataFrame
# 確保 PlayID 和 播映編碼 是整數類型
df1['PlayID'] = df1['PlayID'].astype(int)
df2['播映編碼'] = df2['播映編碼'].astype(int)

# 使用 inner join 合併數據，只保留匹配的記錄
df1 = df1.merge(df2[['播映編碼', '該日票房']], left_on='PlayID', right_on='播映編碼', how='inner')

# 刪除合併後多餘的 '播映編碼' 列
df1.drop('播映編碼', axis=1, inplace=True)

# 將 '該日票房' 更名為 'Daily_box_office'
df1.rename(columns={'該日票房': 'Daily_box_office'}, inplace=True)
```

```python
# 現在 df1 僅包含匹配 '播映編碼' 的 'Daily_box_office' 數據
print(df1['Daily_box_office'])

# 檢查數據信息
print(df1.info())

# 檢查缺失值
print(df1.isnull().sum())

# 處理缺失值
# data.fillna(method='ffill', inplace=True)  # 以前一個數值填充，也可選擇其他方法
```

```python
# 假設 df1 已經是合併後的數據框
# 將 df1 寫入到 Excel 文件中，文件名為 'merged_data.xlsx'
df1.to_excel(r"C:\Users\User\Dropbox\Movie\df3.xlsx", index=False)
```
