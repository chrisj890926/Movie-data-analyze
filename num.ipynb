## 針對數值變數

```python
import pandas as pd

# 載入數據
df1 = pd.read_excel(r'C:\Users\User\Dropbox\Movie\df3.xlsx')
```

```python
numerical_vars = [
    'Daily_box_office', 'Audience', 'Tempture', 'Rainfall', 
    'Movie_Length', 'Budget', 'IMDb_rating', 'Director_Award', 
    'Actor_Awardall', 'Subsidy_Amount', 'Oscar_AwardTotal', 
    'Cannes_AwardTotal', 'TFF_Nom', 'TFF_AwardTotal', 'Distrbutor'
]
```

```python
df2 = df1[numerical_vars].copy()
```

```python
print(df2.head())  # 查看前几行数据
print(df2.info())  # 获取 DataFrame 的基本信息
```

```python
# 获取描述性统计信息
descriptive_stats = df2.describe()
print(descriptive_stats)
```

```python
# 计算偏度
skewness = df2.skew()
print("Skewness:\n", skewness)

# 计算峰度
kurtosis = df2.kurtosis()
print("Kurtosis:\n", kurtosis)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 绘制直方图
df2.hist(bins=15, figsize=(15, 10), layout=(4, 4))
plt.tight_layout()
plt.show()

# 绘制箱形图
plt.figure(figsize=(15, 10))
df2.plot(kind='box', subplots=True, layout=(4, 4), sharex=False, sharey=False)
plt.tight_layout()
plt.show()
```
